# Lab 13 - Linear Regression

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

### Read Health insurance data and display 5 first rows

```python
df = pd.read_csv('./data/car_data.csv')
df.head()
```

### Get data description and info

```python
df.describe
```

```python
df.info
```

## Hypothesis 1 - The newer the car is the more it costs

### Define dependent and independent variables

```python
X = df["Year"].values
y = df["Selling_Price"].values

print('X =', X)
print('y =', y)
```

### Reshape the data of X

```python
X = X.reshape(-1, 1)
print('X =', X[:20])
```

### Get train and test data (80/20)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=100)
print('X_train =', X_train[:20])
print('y_train =', y_train[:20])
```

### Plot the data

```python
plt.scatter(X_train, y_train, color='red')
plt.xlabel("Year of manufacture")
plt.ylabel("Car price")
plt.title("The newer the car is the higher is the price")
plt.show()
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
y_predict = lm.predict(X_test)
print(f"Train accuracy {round(lm.score(X_train, y_train) * 100,2)} %")
print(f"Test accuracy {round(lm.score(X_test, y_test) * 100,2)} %")
```

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_test, y_predict)
plt.xlabel("Year of manufacture")
plt.ylabel("Car price")
plt.title("The newer the car is the higher is the price")
plt.show()
```
